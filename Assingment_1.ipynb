Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.decomposition import LatentDirichletAllocation, NMF, TruncatedSVD
```

Load Dataset

```python
df = pd.read_csv("/content/movie_reviews_dataset.csv")
print(df.head())
```

text Preprocessing

```python
import re

def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^a-zA-Z]', ' ', text)
    return text

df['clean_text'] = df['text'].apply(clean_text)
```

Vectorization

```python
cv = CountVectorizer(stop_words='english')
dtm = cv.fit_transform(df['clean_text'])
```

```python
tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(df['clean_text'])
```

lda

```python
lda = LatentDirichletAllocation(n_components=3, random_state=42)
lda.fit(dtm)
```

nmf

```python
nmf = NMF(n_components=3, random_state=42)
nmf.fit(tfidf_matrix)
```

LSA

```python
lsa = TruncatedSVD(n_components=3, random_state=42)
lsa.fit(tfidf_matrix)
```

```python
def show_topics(model, feature_names, n_words=5):
    for i, topic in enumerate(model.components_):
        print(f"\nTopic {i}:")
        print([feature_names[j] for j in topic.argsort()[-n_words:]])

print("LDA Topics:")
show_topics(lda, cv.get_feature_names_out())

print("\nNMF Topics:")
show_topics(nmf, tfidf.get_feature_names_out())

print("\nLSA Topics:")
show_topics(lsa, tfidf.get_feature_names_out())
```

```python
lda_output = lda.transform(dtm)
df['Topic'] = lda_output.argmax(axis=1)

print(df[['text', 'Topic']])
```
