Import Libraries

```python
!pip install gensim
import pandas as pd
import re
from gensim.models import Word2Vec
from nltk.tokenize import word_tokenize
import nltk

nltk.download('punkt')
```

Load Dataset

```python
df = pd.read_csv('/content/movie_reviews_dataset.csv')
print(df.head())
```

Preprocessing

```python
nltk.download('punkt_tab')
def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^a-zA-Z]', ' ', text)
    return text

df['clean_text'] = df['text'].apply(clean_text)

# Tokenization
corpus = [word_tokenize(text) for text in df['clean_text']]
```

CBOW Mode

```python
cbow_model = Word2Vec(
    sentences=corpus,
    vector_size=100,
    window=5,
    min_count=1,
    sg=0
)
```

```python
skip_model = Word2Vec(
    sentences=corpus,
    vector_size=100,
    window=5,
    min_count=1,
    sg=1
)
```

Vocabulary Size

```python
print("CBOW Vocabulary Size:", len(cbow_model.wv))
print("Skip-Gram Vocabulary Size:", len(skip_model.wv))
```

Word Vectors




```python
words = ["movie", "acting", "story", "boring", "great"]

for word in words:
    print(f"\nWord: {word}")
    print("CBOW:", cbow_model.wv[word][:5])
    print("Skip-Gram:", skip_model.wv[word][:5])
```

```python
word = "movie"

print("\nCBOW Similar Words:")
print(cbow_model.wv.most_similar(word, topn=5))

print("\nSkip-Gram Similar Words:")
print(skip_model.wv.most_similar(word, topn=5))
```
