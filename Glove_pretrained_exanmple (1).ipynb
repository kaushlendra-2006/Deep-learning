```python
import gensim
import numpy as np
from numpy.linalg import norm
import gensim.downloader as apl

glove=apl.load("glove-wiki-gigaword-50")
```

```python
similarity1=glove.similarity("king","queen")
similarity2=glove.similarity("king","apple")

print("similarity(king,queen):",similarity1)
print("similarity(king,apple):",similarity2)
```

```python
def sentence_vector (sentence):
  words=sentence.lower().split()
  vectors=[]

  for word in words:
    if word in glove:
      vectors.append(glove[word])
  # Handle the case where no words from the sentence are found in the glove model
  if vectors:
    return np.mean(vectors,axis=0)
  else:
    # Return a zero vector of the appropriate size if no words are found
    return np.zeros(glove.vector_size)
```

```python
sentence1="i love machine learning"
sentence2="i love deep learning"
sentence3="The cat is sleeping on the mat"

vector1=sentence_vector(sentence1)
vector2=sentence_vector(sentence2)
```
