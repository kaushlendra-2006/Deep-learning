```python
stop_words = set(stopwords.words('english'))

def preprocess(text):
    text = str(text).lower()
    text = re.sub(r'\W', ' ', text)   # remove special chars
    tokens = word_tokenize(text)

    tokens = [word for word in tokens if word not in stop_words and len(word) > 3]

    return tokens

df['processed_text'] = df['Content'].apply(preprocess)
```

```python
dictionary = corpora.Dictionary(df['processed_text'])

# Remove rare/common words
dictionary.filter_extremes(no_below=5, no_above=0.5)

corpus = [dictionary.doc2bow(text) for text in df['processed_text']]
```

```python
lda_model = LdaModel(
    corpus=corpus,
    id2word=dictionary,
    num_topics=5,   # you can change (5–10)
    random_state=42,
    passes=10
)
```

```python
for idx, topic in lda_model.print_topics(-1):
    print(f"Topic {idx}: {topic}\n")
```

```python
doc_topics = lda_model.get_document_topics(corpus)

print(doc_topics[0])
```

```python
vis = pyLDAvis.gensim_models.prepare(lda_model, corpus, dictionary)
pyLDAvis.show(vis)
```

```python
def get_main_topic(doc):
    return sorted(lda_model.get_document_topics(doc), key=lambda x: x[1], reverse=True)[0][0]

df['topic'] = [get_main_topic(doc) for doc in corpus]

df.head()
```
