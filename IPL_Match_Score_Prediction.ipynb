Project Objective

The goal of this project is to predict the cricket match score using deep learning models:

RNN
LSTM
GRU

## The model learns from previous ball-by-ball match data and predicts the cumulative score.

## Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, SimpleRNN, LSTM, GRU
```

## Load Dataset

```python
df = pd.read_csv("/content/deliveries.csv")

print(df.head())
```

## Data Preprocessing

```python
# Select useful columns
data = df[['match_id', 'inning', 'over', 'ball',
           'total_runs']]

# Create cumulative score
data['cumulative_runs'] = data.groupby(
    ['match_id', 'inning']
)['total_runs'].cumsum()

print(data.head())
```

Create Feature

```python
data_cleaned = data.dropna()
X = data_cleaned[['inning', 'over', 'ball']]
y = data_cleaned['cumulative_runs']
```

## Feature Scaling

```python
scaler = MinMaxScaler()

X_scaled = scaler.fit_transform(X)
```

## Reshape Data for RNN/LSTM/GRU

```python
X_scaled = X_scaled.reshape(
    (X_scaled.shape[0], 1, X_scaled.shape[1])
)
```

## Train Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled,
    y,
    test_size=0.2,
    random_state=42
)
```

## Build Simple RNN Model

```python
rnn_model = Sequential()

rnn_model.add(SimpleRNN(
    64,
    activation='relu',
    input_shape=(1, 3)
))

rnn_model.add(Dense(1))

rnn_model.compile(
    optimizer='adam',
    loss='mse'
)

rnn_model.fit(
    X_train,
    y_train,
    epochs=10,
    batch_size=32
)
```

## Build LSTM Model

```python
lstm_model = Sequential()

lstm_model.add(LSTM(
    64,
    activation='relu',
    input_shape=(1, 3)
))

lstm_model.add(Dense(1))

lstm_model.compile(
    optimizer='adam',
    loss='mse'
)

lstm_model.fit(
    X_train,
    y_train,
    epochs=10,
    batch_size=32
)
```

## **Build GRU Model**

```python
gru_model = Sequential()

gru_model.add(GRU(
    64,
    activation='relu',
    input_shape=(1, 3)
))

gru_model.add(Dense(1))

gru_model.compile(
    optimizer='adam',
    loss='mse'
)

gru_model.fit(
    X_train,
    y_train,
    epochs=10,
    batch_size=32
)
```

## Predictions

```python
rnn_pred = rnn_model.predict(X_test)
lstm_pred = lstm_model.predict(X_test)
gru_pred = gru_model.predict(X_test)
```

## Accuracy Comparison

```python

rnn_mae = mean_absolute_error(y_test, rnn_pred)
lstm_mae = mean_absolute_error(y_test, lstm_pred)
gru_mae = mean_absolute_error(y_test, gru_pred)

print("RNN MAE:", rnn_mae)
print("LSTM MAE:", lstm_mae)
print("GRU MAE:", gru_mae)
```

## Visualization

```python
models = ['RNN', 'LSTM', 'GRU']
errors = [rnn_mae, lstm_mae, gru_mae]

plt.figure(figsize=(8,5))

plt.bar(models, errors)

plt.title("Model Comparison")
plt.ylabel("MAE Error")

plt.show()
```
